```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.set_theme()
sns.scatterplot(x=df["total_bill"], y=df["tip"]);
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(12, 9))
positions = [(0, 0), (0, 1), (1, 0), (1, 1)]

for context, (x, y) in zip(["paper", "notebook", "talk", "poster"], positions):
    sns.set_theme(context=context)
    ax = axs[x][y]
    sns.scatterplot(x=df["total_bill"], y=df["tip"], ax=ax)
    ax.set_title(f"sns.set_theme(context={context!r})")
```

```python
fig = plt.figure(figsize=(15, 10))
styles = ["whitegrid", "white", "ticks",  "darkgrid", "dark"]
positions1 = [231, 232, 233, 234, 235]

for style, position in zip(styles, positions1):
    with sns.axes_style(style):
        sns.set_context("notebook")
        ax1 = fig.add_subplot(position)
        ax1.set_title(f"sns.set_theme(style={style!r})")
        sns.scatterplot(x=df["total_bill"], y=df["tip"], ax=ax1)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, sharex=True, sharey=True, figsize=(15, 9))
palettes = ["Accent", "autumn", "hot", "copper", "flag", "ocean"]
positions = [(0, 0), (0, 1), (0, 2), (1, 0), (1, 1), (1, 2)]

for palette, (x, y) in zip(palettes, positions):
    sns.set_theme(palette=palette)
    ax = axs[x][y]
    sns.violinplot(x="day", y="total_bill", data=df, ax=ax)
    ax.set_title(f"sns.set_theme(palette={palette!r})")
```

```python
sns.set_theme(context="talk", style="ticks", palette="Accent")
plt.figure(figsize=(10, 6))
sns.violinplot(x="day", y="total_bill", data=df);
```

```python
sns.set_theme(context="talk", style="ticks", palette="Accent")
plt.figure(figsize=(10, 6))
ax = sns.violinplot(x="day", y="total_bill", data=df)
ax.set_title("Bill Amount by Day")
ax.set_xlabel("Day of the Week")
ax.set_ylabel("Total Bill Amount in USD")
plt.savefig("tips_violin.png")
```
